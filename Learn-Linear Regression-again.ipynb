```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
boston = datasets.load_boston()
X = boston.data
y = boston.target
```

```python
print(X.shape)
print(X.dtype)
print(y.shape)
print(y.dtype)
```

```python
X
```

```python
plt.plot(range(len(y)),y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 19)

print(X_train.shape)
print(X_train.dtype)
print(y_train.shape)
print(y_train.dtype)
```

```python
from sklearn.preprocessing import StandardScaler
standard = StandardScaler()
standard.fit(X_train)
X_train_transform = standard.transform(X_train)
X_test_transform = standard.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train_transform,y_train)
```

```python
y_prediction = lin_reg.predict(X_test_transform)
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
mse = mean_squared_error(y_test,y_prediction)
rmse = np.sqrt(mean_squared_error(y_test,y_prediction))
mae = mean_absolute_error(y_test,y_prediction)
r2 = r2_score(y_test,y_prediction)

print(mse)
print(rmse)
print(mae)
print(r2)
```

```python
plt.plot(range(len(y_test)),y_test,color = 'r',label = 'y_test')
plt.plot(range(len(y_prediction)),y_prediction,color = 'b',label = 'y_prediction')
```
