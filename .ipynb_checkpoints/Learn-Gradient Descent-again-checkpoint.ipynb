```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,5,150)
y = (x+2.5)**2 + 7
```

```python
plt.plot(x,y)
plt.show()
```

```python
def J(theta):
    return (theta+2.5)**2+7

def dJ(theta):
    return 2*(theta + 2.5)

def gradient_descent(lr , theta):
    limit = 1e-7
    history_theta = [theta]
    while True:
        last_theta = theta
        theta = theta - lr * (dJ(theta))
        history_theta.append(theta)
        y = abs((J(last_theta)-J(theta)))
        if(y < limit):
            break
    return history_theta,  theta  
```

```python
lr = 0.1
theta = -8
history_theta,theta = gradient_descent(lr,theta)
print(J(theta))
print(theta)
```

```python
plt.plot(x,y)
plt.plot(np.array(history_theta),J(np.array(history_theta)),'b+')
```

```python
print(lr)
```

```python
print(len(history_theta))
```

```python
lr = 0.08
theta = -8
history_theta,theta = gradient_descent(lr,theta)
print(J(theta))
print(theta)
plt.plot(x,y)
plt.plot(np.array(history_theta),J(np.array(history_theta)),'b+')
print(len(history_theta))
```

```python
lr = 0.8
theta = -8
history_theta,theta = gradient_descent(lr,theta)
print(J(theta))
print(theta)
plt.plot(x,y)
plt.plot(np.array(history_theta),J(np.array(history_theta)),'r')
print(len(history_theta))
```
